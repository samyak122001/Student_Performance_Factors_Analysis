# Exploratory Data Analysis

# Hours_Studied

```python
df.columns
```

```python
ax= sns.countplot(x= 'Hours_Studied', data= df)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
hrs_studied= df.groupby(['Gender'], as_index=False)['Hours_Studied'].sum().sort_values(by='Hours_Studied', ascending= False)

sns.barplot(x='Gender', y='Hours_Studied', data=  hrs_studied)
```

```python
hrs_studied= df.groupby(['Gender'], as_index=False)['Hours_Studied'].sum().sort_values(by='Hours_Studied', ascending= False)

sns.barplot(x='Gender', y='Hours_Studied', data=  hrs_studied)

```

From the above graphs we can see that and most students studying between 15 and 25 hours and most of the hours studied by Males

# Attendance Rate 

```python
ax= sns.countplot(x= 'Attendance', data= df)
```

```python
Att_Rates = df.groupby(['Gender'], as_index=False)['Attendance'].sum().sort_values(by='Attendance', ascending= False)

sns.barplot(x='Gender', y='Attendance', data= Att_Rates)
```

```python
sns.barplot(x='Hours_Studied', y='Attendance', data= df)
```

Most students have high attendance rates, clustering around 80% to 100% and males have more attendance than females and Students with 2 hours study times has the highest Attendance.

# Sleeping Hours 

```python
df.columns
```

```python
ax= sns.countplot(x= 'Sleep_Hours', data= df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sleep_hrs= df.groupby(['Gender'], as_index=False)['Sleep_Hours'].sum().sort_values(by='Sleep_Hours', ascending= False)

sns.barplot(x='Gender', y='Sleep_Hours', data = sleep_hrs)
```

```python
sleep_hrs= df.groupby(['Attendance'], as_index=False)['Sleep_Hours'].sum().sort_values(by='Sleep_Hours', ascending= False)

sns.barplot(x='Attendance', y='Sleep_Hours', data = sleep_hrs)
```

```python
sleep_hrs= df.groupby(['Hours_Studied'], as_index=False)['Sleep_Hours'].sum().sort_values(by='Sleep_Hours', ascending= False)

sns.barplot(x='Hours_Studied', y='Sleep_Hours', data = sleep_hrs)
```

From the above graph we can see that the majority of students sleep 6 to 8 hours, with a slight peak around 7 hours and male has more sleeping hours than females and student who studied 20 hours has highest sleeping hours.

# Tutoring Sessions

```python
df.columns
```

```python
ax= sns.countplot(x= 'Tutoring_Sessions', data= df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Tutoring_session= df.groupby(['Tutoring_Sessions'], as_index=False)['Exam_Score'].sum().sort_values(by='Exam_Score', ascending= False)

ax= sns.barplot(x='Tutoring_Sessions', y='Exam_Score', data = Tutoring_session)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Tutoring_session= df.groupby(['School_Type'], as_index=False)['Tutoring_Sessions'].sum().sort_values(by='Tutoring_Sessions', ascending= False)

ax= sns.barplot(x='School_Type', y='Tutoring_Sessions', data = Tutoring_session)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Tutoring_session= df.groupby(['Family_Income'], as_index=False)['Tutoring_Sessions'].sum().sort_values(by='Tutoring_Sessions', ascending= False)

ax= sns.barplot(x='Family_Income', y='Tutoring_Sessions', data = Tutoring_session)
for bar in ax.containers:
    ax.bar_label(bar)
```

From the above graphs we can see that student who are taking 1 tutoring session have highest exam score, students who are studying in Public School are  taking  more tutoring sessions, students with medium family income taking more tutoring sessions. 

# Conclusion:

Male students who studied between 15 and 20 hours and had more attendance rates clustering around 80-100% also tended to sleep 6-8 hours, while students with fewer study hours exhibited higher attendance and the highest sleep hours, around 20 hours, were noted.

The scores are broadly distributed, with a significant number of students scoring between 60 and 90. Students with medium parental involvement, access to resources, and motivation levels tend to achieve the highest exam scores, particularly those involved in extracurricular activities and with internet access. Additionally, students taking 1 tutoring session have the highest exam scores, while those in public schools and with medium family incomes are more likely to take additional tutoring sessions.
